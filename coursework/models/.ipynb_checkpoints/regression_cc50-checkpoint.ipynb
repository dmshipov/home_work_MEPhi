Базовые модели для CC50

```python
# Импорты
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Загрузка данных
data = pd.read_csv('/content/chem_cleaned_data.csv')
X = data.drop(columns=['IC50', 'CC50', 'SI', 'Unnamed: 0'], errors='ignore')
y = data['CC50']

# Стандартизация
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

Обучение и сравнение моделей

```python
# Ridge
ridge_params = {'alpha': [0.1, 1.0, 10.0, 100.0]}
ridge_model = GridSearchCV(Ridge(), ridge_params, cv=5, scoring='neg_mean_squared_error')
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.best_estimator_.predict(X_test)

# Lasso
lasso_params = {'alpha': [0.1, 1.0, 10.0]}
lasso_model = GridSearchCV(Lasso(max_iter=70000), lasso_params, cv=5, scoring='neg_mean_squared_error')
lasso_model.fit(X_train, y_train)
y_pred_lasso = lasso_model.best_estimator_.predict(X_test)

# Random Forest
rf_params = {'n_estimators': [50, 100], 'max_depth': [None, 10, 20]}
rf_model = GridSearchCV(RandomForestRegressor(random_state=42), rf_params, cv=3, scoring='neg_mean_squared_error')
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.best_estimator_.predict(X_test)

# XGBoost
xgb_params = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5],
    'learning_rate': [0.05, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
xgb_model = GridSearchCV(
    XGBRegressor(objective='reg:squarederror', random_state=42),
    xgb_params,
    cv=3,
    scoring='neg_mean_squared_error'
)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.best_estimator_.predict(X_test)
```

3. Оценка качества

```python
def evaluate_model(y_true, y_pred, name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f'{name}:\n  RMSE = {rmse:.2f}\n  MAE  = {mae:.2f}\n  R^2  = {r2:.4f}\n')

evaluate_model(y_test, y_pred_ridge, 'Ridge (CC50)')
evaluate_model(y_test, y_pred_lasso, 'Lasso (CC50)')
evaluate_model(y_test, y_pred_rf, 'Random Forest (CC50)')
evaluate_model(y_test, y_pred_xgb, 'XGBoost (CC50)')
```

Отбор топ-50 признаков по важности XGBoost

```python
# Обучаем XGBoost на всех признаках
xgb_full = XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_full.fit(X_train, y_train)

# Получаем важность признаков
importances_cc50 = pd.Series(xgb_full.feature_importances_, index=X.columns)
top_50_cc50 = importances_cc50.sort_values(ascending=False).head(50).index

# Новый X только с топ-50
X_top_cc50 = data[top_50_cc50]
X_top_scaled_cc50 = scaler.fit_transform(X_top_cc50)

# Новый train/test split
X_train_top, X_test_top, y_train_top, y_test_top = train_test_split(
    X_top_scaled_cc50, y, test_size=0.2, random_state=42
)
```

Удаление сильно коррелирующих признаков

```python
corr_matrix = pd.DataFrame(X_top_cc50, columns=top_50_cc50).corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.95)]

# Удаляем коррелирующие признаки
X_top_refined_cc50 = pd.DataFrame(X_top_cc50, columns=top_50_cc50).drop(columns=to_drop)
X_top_refined_scaled = scaler.fit_transform(X_top_refined_cc50)

# Обновлённые train/test
X_train_refined, X_test_refined, y_train_refined, y_test_refined = train_test_split(
    X_top_refined_scaled, y, test_size=0.2, random_state=42
)
```

Обучение XGBoost на отобранных признаках

```python
xgb_cc50_final = XGBRegressor(
    n_estimators=150,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.9,
    colsample_bytree=0.8,
    gamma=1,
    min_child_weight=3,
    objective='reg:squarederror',
    random_state=42
)
xgb_cc50_final.fit(X_train_refined, y_train_refined)
y_pred_cc50_final = xgb_cc50_final.predict(X_test_refined)
```

```python
evaluate_model(y_test_refined, y_pred_cc50_final, 'XGBoost (CC50, после оптимизации)')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test_refined, y_pred_cc50_final, alpha=0.5, edgecolors='k')
plt.plot([y_test_refined.min(), y_test_refined.max()],
         [y_test_refined.min(), y_test_refined.max()],
         'r--', linewidth=2)
plt.xlabel('Реальные значения CC50')
plt.ylabel('Предсказанные значения CC50')
plt.title('XGBoost (финальная модель CC50): Реальное vs Предсказанное')
plt.grid(True)
plt.tight_layout()
plt.show()
```
