```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")
```

```python
# Загрузка
data = pd.read_csv('/content/chem_cleaned_data.csv')

# Признаки и цель
X = data.drop(columns=['IC50', 'CC50', 'SI', 'Unnamed: 0'], errors='ignore')
cc50_median = data['CC50'].median()
y = (data['CC50'] > cc50_median).astype(int)

# Баланс классов
print("Классы (CC50 > медианы):", y.value_counts().to_dict())

# Стандартизация
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделение
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
```

```python
# Logistic Regression
log_params = {'C': [0.1, 1, 10]}
log_model = GridSearchCV(LogisticRegression(max_iter=5000), log_params, cv=5, scoring='f1')
log_model.fit(X_train_bal, y_train_bal)
y_pred_log = log_model.best_estimator_.predict(X_test)

# Random Forest
rf_params = {'n_estimators': [100, 200], 'max_depth': [None, 10, 20]}
rf_model = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=3, scoring='f1')
rf_model.fit(X_train_bal, y_train_bal)
y_pred_rf = rf_model.best_estimator_.predict(X_test)

# XGBoost
xgb_params = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.05, 0.1]
}
xgb_model = GridSearchCV(
    XGBClassifier(eval_metric='logloss', random_state=42),
    xgb_params,
    cv=3,
    scoring='f1'
)
xgb_model.fit(X_train_bal, y_train_bal)
y_pred_xgb = xgb_model.best_estimator_.predict(X_test)
```

```python
def evaluate_classifier(y_true, y_pred, name):
    print(f"\n{name}")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("F1-score:", f1_score(y_true, y_pred))
    print("ROC AUC:", roc_auc_score(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))

evaluate_classifier(y_test, y_pred_log, "Logistic Regression (CC50 > медианы)")
evaluate_classifier(y_test, y_pred_rf, "Random Forest (CC50 > медианы)")
evaluate_classifier(y_test, y_pred_xgb, "XGBoost (CC50 > медианы)")
```

```python
log_proba = log_model.best_estimator_.predict_proba(X_test)[:, 1]
rf_proba = rf_model.best_estimator_.predict_proba(X_test)[:, 1]
xgb_proba = xgb_model.best_estimator_.predict_proba(X_test)[:, 1]

plt.figure(figsize=(8, 6))
for name, y_score in [
    ("Logistic", log_proba),
    ("Random Forest", rf_proba),
    ("XGBoost", xgb_proba)
]:
    fpr, tpr, _ = roc_curve(y_test, y_score)
    auc = roc_auc_score(y_test, y_score)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.3f})")

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривые моделей (CC50 > медианы)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
