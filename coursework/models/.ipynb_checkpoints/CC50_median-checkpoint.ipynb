## Введение

Задача классификации, определяющая, превышает ли значение СС50 медианное значение выборки. В рамках данного анализа будут использованы различные методы машинного обучения для решения данной задачи, а также выполнена оценка их производительности и интерпретация результатов.

```python
# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")
```

```python
# Добавим данные в датафрейм
df = pd.read_csv("data.csv")
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# Проверим загрузку данных
df.head(10)
```

```python
# Целевая переменная
target = 'CC50'

# Отделяем признаки от целевой переменной
X = df.drop(columns=[target, 'CC50', 'SI'], errors='ignore')
CC_median = df[target].median()
y = (df[target] > CC_median).astype(int)
```

```python
# Стандартизация признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Проверяем баланс классов
print("Классы (CC50 > медианы):", y.value_counts().to_dict())

# Визуализируем балансировку классов
plt.figure(figsize=(12, 6))
sns.countplot(x=y_train, palette="Set2")
plt.title("Распределение классов")
plt.xlabel('Класс')
plt.ylabel('Количество наблюдений')
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
# Напишем функцию для оценки моделей
def evaluate(y_true, y_pred, model_name):
    print(f"\nРезультаты для модели: {model_name}")
    print("Точность:", accuracy_score(y_true, y_pred))
    print("F1-score:", f1_score(y_true, y_pred))
    print("ROC AUC:", roc_auc_score(y_true, y_pred))
    print("Матрица ошибок:\n", confusion_matrix(y_true, y_pred))
```

```python
# Логистическая регрессия
log_params = {'C': [0.05, 0.5, 1.5, 5]}
log_model = GridSearchCV(LogisticRegression(max_iter=7000, solver='liblinear'), log_params, cv=4, scoring='f1')
log_model.fit(X_train, y_train)
y_pred_log = log_model.best_estimator_.predict(X_test)

# Оценим производительность моделей
evaluate(y_test, y_pred_log, "Логистическая Регрессия (CC50 > медианы)")

# Вероятности положительных классов
probs = log_model.best_estimator_.predict_proba(X_test)[:, 1]

# Рассчитываем FPR и TPR
fpr, tpr, thresholds = roc_curve(y_test, probs)

# Рассчитаем площадь под кривой (AUC)
roc_auc = roc_auc_score(y_test, probs)

# Рисуем график ROC-кривой
plt.figure(figsize=(8, 6))
lw = 2
plt.plot(fpr, tpr, color='blue',
         lw=lw, label=f'ROC-кривая (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Частота ложных срабатываний')
plt.ylabel('Частота истинных срабатываний')
plt.title('ROC-кривая для Логистическая Регрессия (CC50 > медианы)')
plt.legend(loc="lower right")
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
# Случайный лес
rf_params = {'n_estimators': [150, 250, 300], 'max_depth': [None, 15, 25]  }
rf_model = GridSearchCV(RandomForestClassifier(random_state=123), rf_params, cv=4, scoring='f1')
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.best_estimator_.predict(X_test)

# Оценим производительность моделей
evaluate(y_test, y_pred_rf, "Случайный Лес (CC50 > медианы)")

# Вероятности положительного класса
probs = rf_model.best_estimator_.predict_proba(X_test)[:, 1]

# Рассчитываем FPR и TPR
fpr, tpr, thresholds = roc_curve(y_test, probs)

# Площадь под кривой (AUC)
roc_auc = roc_auc_score(y_test, probs)

# Отрисовка графика ROC-кривой
plt.figure(figsize=(8, 6))
lw = 2
plt.plot(fpr, tpr, color='forestgreen', lw=lw,
         label=f'ROC-кривая (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Частота ложных срабатываний')
plt.ylabel('Частота истинных срабатываний')
plt.title('ROC-кривая для Случайного Леса (CC50 > медианы)')
plt.legend(loc="lower right")
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
# Модель XGBoost
xgb_params = {
    'n_estimators': [150, 250, 350],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.15] 
}
xgb_model = GridSearchCV(
    XGBClassifier(eval_metric='logloss', random_state=123),
    xgb_params,
    cv=4,
    scoring='f1'
)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.best_estimator_.predict(X_test)

# Оценим производительность моделей
evaluate(y_test, y_pred_xgb, "XGBoost (CC50 > медианы)")

# Вероятности положительного класса
probs = xgb_model.best_estimator_.predict_proba(X_test)[:, 1]

# Рассчитываем FPR и TPR
fpr, tpr, thresholds = roc_curve(y_test, probs)

# Площадь под кривой (AUC)
roc_auc = roc_auc_score(y_test, probs)

# Нарисум график ROC-кривой
plt.figure(figsize=(8, 6))
lw = 2
plt.plot(fpr, tpr, color='chocolate', lw=lw, label=f'ROC-кривая (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Частота ложных срабатываний')
plt.ylabel('Частота истинных срабатываний')
plt.title('ROC-кривая для XGBoost (CC50 > медианы)')
plt.legend(loc="lower right")
plt.grid(axis='y', alpha=0.7)
plt.show()
```

Сравним результаты

```python
# Вероятности положительного класса для каждой модели
log_proba = log_model.best_estimator_.predict_proba(X_test)[:, 1]
rf_proba = rf_model.best_estimator_.predict_proba(X_test)[:, 1]
xgb_proba = xgb_model.best_estimator_.predict_proba(X_test)[:, 1]

# Цветовая палитра Seaborn (цвета были выбраны ранее вручную)
colors = ['blue', 'forestgreen', 'chocolate']

# Построение графика
plt.figure(figsize=(10, 7))

# Проходим по каждой модели и строим её ROC-кривую
for name, y_score, color in zip(
    ["Логистическая Регрессия", "Случайный Лес", "XGBoost"],
    [log_proba, rf_proba, xgb_proba],
    colors
):
    fpr, tpr, _ = roc_curve(y_test, y_score)
    auc = roc_auc_score(y_test, y_score)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.3f})", color=color, linewidth=2)

# Добавляем линию случайного угадывания
plt.plot([0, 1], [0, 1], 'k--', label='Случайное угадывание', linewidth=1)

# Подписи и оформление
plt.xlabel('Частота ложных срабатываний', fontsize=14)
plt.ylabel('Частота истинных срабатываний', fontsize=14)
plt.title('ROC-кривые моделей для предсказания CC50', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.tight_layout()

# Показываем график
plt.show()
```
