## Введение

Задача классификации, определяющая, превышает ли значение SI медианное значение выборки. В рамках данного анализа будут использованы различные методы машинного обучения для решения данной задачи, а также выполнена оценка их производительности и интерпретация результатов.

```python
# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")
```

```python
# Добавим данные в датафрейм
df = pd.read_csv('../../data/data.csv')
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# Проверим загрузку данных
df.head(10)
```

```python
# Целевая переменная
target = 'SI'

# Отделяем признаки от целевой переменной
X = df.drop(columns=[target, 'CC50', 'IC50'], errors='ignore')
SI_median = df[target].median()
y = (df[target] > SI_median).astype(int)
```

```python
# Стандартизация признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Проверяем баланс классов
print("Классы (SI > медианы):", y.value_counts().to_dict())

# Визуализируем балансировку классов
plt.figure(figsize=(12, 6))
sns.countplot(x=y_train, palette="Set2")
plt.title("Распределение классов")
plt.xlabel('Класс')
plt.ylabel('Количество наблюдений')
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
# Напишем функцию для оценки моделей
def evaluate(y_true, y_pred, model_name):
    print(f"\nРезультаты для модели: {model_name}")
    print("Точность:", accuracy_score(y_true, y_pred))
    print("F1-score:", f1_score(y_true, y_pred))
    print("ROC AUC:", roc_auc_score(y_true, y_pred))
    print("Матрица ошибок:\n", confusion_matrix(y_true, y_pred))
```

```python
# Логистическая регрессия
log_params = {'C': [0.05, 0.5, 1.5, 5]}
log_model = GridSearchCV(LogisticRegression(max_iter=7000, solver='liblinear'), log_params, cv=4, scoring='f1')
log_model.fit(X_train, y_train)
y_pred_log = log_model.best_estimator_.predict(X_test)


# Вероятности положительных классов
probs = log_model.best_estimator_.predict_proba(X_test)[:, 1]

# Рассчитываем FPR и TPR
fpr, tpr, thresholds = roc_curve(y_test, probs)

# Оценим производительность моделей
evaluate(y_test, y_pred_log, "Логистическая Регрессия (SI > медианы)")

# Рассчитываем матрицу ошибок
cm = confusion_matrix(y_test, y_pred_log)

# Визуализируем матрицу ошибок
plt.figure(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Матрица ошибок для Логистической Регрессии (SI > медианы)')
plt.grid(False)
plt.show()
```

Логистическая регрессия демонстрирует хорошую производительность, достигая высокой точности и F1-score. Однако, присутствие небольшого количества ошибок (ложноположительных и ложноотрицательных) указывает на потенциал для улучшения.

```python
# Случайный лес
rf_params = {'n_estimators': [150, 250, 300], 'max_depth': [None, 15, 25]  }
rf_model = GridSearchCV(RandomForestClassifier(random_state=123), rf_params, cv=4, scoring='f1')
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.best_estimator_.predict(X_test)

# Оценим производительность моделей
evaluate(y_test, y_pred_rf, "Случайный Лес (SI > медианы)")

# Вероятности положительного класса
probs = rf_model.best_estimator_.predict_proba(X_test)[:, 1]

# Рассчитываем FPR и TPR
fpr, tpr, thresholds = roc_curve(y_test, probs)

# Рассчитываем матрицу ошибок
cm = confusion_matrix(y_test, y_pred_rf)

# Визуализируем матрицу ошибок
plt.figure(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Greens)
plt.title('Матрица ошибок для Логистической Регрессии (SI > медианы)')
plt.grid(False)
plt.show()
```

Случайный лес превосходит логистическую регрессию по всем метрикам. Модель демонстрирует более высокую точность и F1-score, а также лучше справляется с разделением классов.

```python
# Модель XGBoost
xgb_params = {
    'n_estimators': [150, 250, 350],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.15] 
}
xgb_model = GridSearchCV(
    XGBClassifier(eval_metric='logloss', random_state=123),
    xgb_params,
    cv=4,
    scoring='f1'
)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.best_estimator_.predict(X_test)

# Оценим производительность моделей
evaluate(y_test, y_pred_xgb, "XGBoost (SI > медианы)")

# Вероятности положительного класса
probs = xgb_model.best_estimator_.predict_proba(X_test)[:, 1]

# Рассчитываем FPR и TPR
fpr, tpr, thresholds = roc_curve(y_test, probs)

# Рассчитываем матрицу ошибок
cm = confusion_matrix(y_test, y_pred_xgb)

# Визуализируем матрицу ошибок
plt.figure(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Reds)
plt.title('Матрица ошибок для Логистической Регрессии (SI > медианы)')
plt.grid(False)
plt.show()
```

XGBoost показывает наилучшие результаты, достигая идеальной точности и F1-score. Модель идеально разделяет классы, что говорит о её превосходной способности выявлять закономерности в данных.

```python
# Прогнозы для каждой модели
log_preds = log_model.best_estimator_.predict(X_test)
rf_preds = rf_model.best_estimator_.predict(X_test)
xgb_preds = xgb_model.best_estimator_.predict(X_test)
# Создаем фигуру для отображения матриц ошибок
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Названия моделей и цветовые палитры
model_names = ["Логистическая Регрессия", "Случайный Лес", "XGBoost"]
predictions = [log_preds, rf_preds, xgb_preds]
colors = ['Blues', 'Greens', 'Reds']

# Проходим по каждой модели и строим матрицу ошибок
for ax, name, preds, color in zip(axes, model_names, predictions, colors):
    cm = confusion_matrix(y_test, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Отрицательный', 'Положительный'])
    disp.plot(ax=ax, cmap=color, values_format='d')
    ax.title.set_text(name)
    
# Общие параметры графика
plt.suptitle('Сравнение моделей по матрицам ошибок', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

XGBoost является лучшей моделью для данной задачи классификации. Она демонстрирует идеальную производительность и должна быть предпочтительной.
Случайный лес показывает хорошие результаты, которые незначительно уступают XGBoost. Эта модель может быть хорошей альтернативой, особенно если требуется более быстрая скорость обучения/предсказания или интерпретируемость.
Логистическая регрессия показывает удовлетворительную производительность, но уступает другим моделям. Она может быть полезна в качестве базовой модели или если требуется простота и скорость.
Рекомендуется использовать модель XGBoost для решения задачи классификации. Если требуется более простая и/или быстрая модель, можно рассмотреть Случайный Лес. Логистическая Регрессия может использоваться для простого анализа или в качестве базовой линии.
