## Введение
Задача предсказания значения СC50 на основе химических дескрипторов является важной для первичного скрининга лекарственных соединений. Для её решения будем использовать различные подходы машинного обучения с оценкой производительности и интерпретацией моделей.

Импоритруем необходимые библиотеки

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, StackingRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.feature_selection import VarianceThreshold
from scipy.stats import randint, uniform

import warnings
warnings.filterwarnings('ignore')
```

Загружаем подготовленные данные

```python
# Добавим данные в датафрейм
df = pd.read_csv('../../data/data.csv')
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# Проверим загрузку данных
df.head(10)
```

Определяем целеую переменную

```python
# Целевая переменная
target = 'Log_CC50'

# Отделяем признаки от целевой переменной
X = df.drop(columns=[target, 'Log_IC50', 'Log_SI'], errors='ignore')
y = df[target]
```

```python
# Проверим на константные признаки и удаление
selector = VarianceThreshold(threshold=0)
selector.fit(X)
constant_features = X.columns[~selector.get_support()]
if len(constant_features) > 0:
    print(f"Удаляем константные признаки: {list(constant_features)}")
    X.drop(columns=constant_features, inplace=True)
```

```python
# Масштабируем признаки
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Разделим данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# Напишем функцию оценки модели
def assess(y_true, y_pred, model_name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"Результаты для модели {model_name}:")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  R2:   {r2:.4f}\n")
```

# Настроем и обученим модель с подбором гиперпараметров

```python
# Ridge-регрессия
ridge_params = {'alpha': [0.1, 1.0, 10.0, 100.0]}
ridge_cv = GridSearchCV(Ridge(), ridge_params, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
ridge_cv.fit(X_train, y_train)
ridge_best = ridge_cv.best_estimator_
ridge_pred = ridge_best.predict(X_test)
assess(y_test, ridge_pred, 'Ridge Regression')

# Визуализация результатов регрессии
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_test, ridge_pred, alpha=0.7, color='blue', edgecolors='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--') 
plt.title('Ridge Regression: Предсказание vs Реальые занчения')
plt.xlabel('Предсказание')
plt.ylabel('Реальные значения')
plt.grid()
```

**Ridge Regression продемонстрировала удовлетворительные результаты, объясняя 78.75% вариации целевой переменной, однако показатели RMSE и MAE указывают на значительные ошибки в предсказаниях.**

```python
# Lasso-регрессия
lasso_params = {'alpha': np.logspace(-4, 0, 10)}
lasso_cv = GridSearchCV(Lasso(max_iter=15000), lasso_params, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
lasso_cv.fit(X_train, y_train)
lasso_best = lasso_cv.best_estimator_
lasso_pred = lasso_best.predict(X_test)
assess(y_test, lasso_pred, 'Lasso Regression')

# Визуализация результатов регрессии
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_test, lasso_pred, alpha=0.7, color='blue', edgecolors='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title('Lasso Regression: Предсказание vs Реальые занчения')
plt.xlabel('Предсказание')
plt.ylabel('Реальные значения')
plt.grid()
```

**Результаты Lasso Regression схожи с Ridge Regression, показывая немного лучшие значения R² (78.95%) и чуть меньшую RMSE, что указывает на небольшое улучшение в точности, но в целом разница незначительна.**

```python
# Случайный лес
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
rf_cv = GridSearchCV(RandomForestRegressor(random_state=42), rf_params, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
rf_cv.fit(X_train, y_train)
rf_best = rf_cv.best_estimator_
rf_pred = rf_best.predict(X_test)
assess(y_test, rf_pred, 'Random Forest')

# Визуализация результатов регрессии
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_test, rf_pred, alpha=0.7, color='blue', edgecolors='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title('Random Forest: Предсказание vs Реальные значения')
plt.xlabel('Предсказание')
plt.ylabel('Реальные значения')
plt.grid()
```

**Random Forest значительно улучшает точность предсказаний, с R², близким к 1, что означает, что модель практически полностью объясняет вариацию целевого показателя. Ошибки RMSE и MAE также минимальны, что указывает на очень высокую точность модели.**

```python
# XGBoost с RandomizedSearchCV
xgb_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 25, 47],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 1, 5]
}
xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42, verbosity=0)
xgb_rand = RandomizedSearchCV(xgb_model, xgb_params, n_iter=25, cv=3, scoring='neg_mean_squared_error', random_state=88, n_jobs=-1)
xgb_rand.fit(X_train, y_train)
xgb_best = xgb_rand.best_estimator_
xgb_pred = xgb_best.predict(X_test)
assess(y_test, xgb_pred, 'XGBoost')

# Визуализация результатов регрессии
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_test, xgb_pred, alpha=0.7, color='blue', edgecolors='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title('XGBoost: Предсказание vs Реальные значения')
plt.xlabel('Предсказание')
plt.ylabel('Реальные значения')
plt.grid()
```

**XGBoost также демонстрирует выдающиеся результаты, с R² равным 99.98%. Хотя RMSE и MAE выше, чем у Random Forest, в целом XGBoost показывает очень высокую точность, оказываясь чуть менее оптимальной по сравнению с Random Forest.**

**Перед подбором гиперпараметров для XGBoost был проведен отбор признаков и исключены те, у которых коэффициент корреляции превышал 0.94.**

# Анализ мультиколлинеарности среди признаков 

```python
# Корреляционная матрица признаков
corr_matrix = pd.DataFrame(X_scaled, columns=X.columns).corr().abs()

# Верхний треугольник матрицы корреляций
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Список признаков для удаления
to_remove = [col for col in upper_tri.columns if any(upper_tri[col] > 0.94)]

print(f"Признаки, которые удалим из-за высокой корреляции (>0.94): {to_remove}")

# Удаляем сильно коррелированные признаки из исходных данных
X_reduced = X.drop(columns=to_remove)
X_reduced_scaled = scaler.fit_transform(X_reduced)

# Делим заново
X_train_red, X_test_red, y_train_red, y_test_red = train_test_split(
    X_reduced_scaled, y, test_size=0.2, random_state=42
)
```

**При анализе данных были идентифицированы и удалены признаки, обладающие высокой корреляцией (более 0.94).**

**Удаление этих признаков помогает снизить влияние мультиколлинеарности, что позволяет моделям учиться более эффективно и улучшает их предсказательные способности.**

Проверим результаты после удаления признаков обладающих высокой корреляцией

```python
# Определяем пространство гиперпараметров для поиска
param_dist = {
    'n_estimators': randint(1, 600),
    'max_depth': randint(1, 15),
    'learning_rate': uniform(0.01, 0.5),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'gamma': uniform(0, 5),
    'reg_alpha': uniform(0, 1),
    'reg_lambda': uniform(0, 1)
}


# Создаем RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=25,
    scoring='accuracy',   
    cv=3,                
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Запускаем подбор гиперпараметров на уменьшенном наборе признаков
random_search.fit(X_train_red, y_train_red)

print("Лучшие параметры:", random_search.best_params_)

# Обучаем модель с лучшими параметрами
xgb_best = random_search.best_estimator_
xgb_red_pred = xgb_best.predict(X_test_red)

assess(y_test_red, xgb_red_pred, 'XGBoost после удаления мультиколлинеарных признаков с автоподбором гиперпараметров')

# Визуализация результатов регрессии
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_test_red, xgb_red_pred, alpha=0.7, color='blue', edgecolors='k')
plt.plot([y_test_red.min(), y_test_red.max()], [y_test_red.min(), y_test_red.max()], 'r--')
plt.title('XGBoost:  Предсказание vs Реальные значения (После обработки)')
plt.xlabel('Предсказание')
plt.ylabel('Реальные значения')
plt.grid()
```

Удаление признаков обладающих высокой корреляцией и подбор гиперпараметров не улучшило результаты, а наоборот ухудшило их, что ожидаемо так как они идеальны и сложно их улучшить

# Визуализация признаков

```python
feat_importances = pd.Series(xgb_best.feature_importances_, index=X_reduced.columns)
feat_importances = feat_importances.sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6))
sns.barplot(x=feat_importances.values, y=feat_importances.index, palette='viridis')
plt.title('Топ-20 важных признаков по XGBoost')
plt.xlabel('Важность признака')
plt.ylabel('Признак')
plt.tight_layout()
plt.show()
```

## Вывод

`Random Forest` и `XGBoost` продемонстрировали значительно лучшие результаты по сравнению с `Ridge` и `Lasso` регрессиями, что свидетельствует о более высокой эффективности этих алгоритмов в данной задаче.

**В данном случае стоит предпочесть использование Random Forest и XGBoost для задач регрессии, так как они показывают наиболее высокие результаты по всем метрикам.**
