## –ó–∞–¥–∞—á–∏

1. –ü—Ä–æ–≤–µ—Å—Ç–∏ –ø–æ–¥—Ä–æ–±–Ω—ã–π —Ä–∞–∑–≤–µ–¥–æ—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö (EDA), –∏—Å–ø–æ–ª—å–∑—É—è —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–æ–Ω–Ω—ã–µ –º–µ—Ç–æ–¥—ã –¥–ª—è –ø–æ–Ω–∏–º–∞–Ω–∏—è —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –∏ —Å–≤–æ–π—Å—Ç–≤ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö.
2. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ —Å—Ä–∞–≤–Ω–∏—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª–µ–π –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è –¥–ª—è —Ä–µ—à–µ–Ω–∏—è —Å–ª–µ–¥—É—é—â–∏—Ö –∑–∞–¥–∞—á:
    - –†–µ–≥—Ä–µ—Å—Å–∏—è –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π IC50, CC50 –∏ SI.
    - –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É –ø—Ä–µ–≤—ã—à–µ–Ω–∏—è –º–µ–¥–∏–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π IC50, CC50 –∏ SI.
    - –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É –ø—Ä–µ–≤—ã—à–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏—è SI –ø–æ—Ä–æ–≥–∞ 8.
3. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –º–æ–¥–µ–ª–µ–π, —Å—Ä–∞–≤–Ω–∏—Ç—å –∏—Ö –∫–∞—á–µ—Å—Ç–≤–æ –∏ —Å–¥–µ–ª–∞—Ç—å –æ–±–æ—Å–Ω–æ–≤–∞–Ω–Ω—ã–µ –≤—ã–≤–æ–¥—ã –æ –Ω–∞–∏–±–æ–ª–µ–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–¥—Ö–æ–¥–∞—Ö.

## –í–∞–∂–Ω–æ—Å—Ç—å –∑–∞–¥–∞—á–∏

–û—Ç –∫–∞—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–≤–µ–¥—ë–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞–ø—Ä—è–º—É—é –∑–∞–≤–∏—Å–∏—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∏ –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å —Ä–∞–∑—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º—ã—Ö –ª–µ–∫–∞—Ä—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–µ–ø–∞—Ä–∞—Ç–æ–≤. –ü–æ—ç—Ç–æ–º—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ç—â–∞—Ç–µ–ª—å–Ω–æ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ, –≤—ã–±—Ä–∞—Ç—å —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏, —É—Å—Ç—Ä–∞–Ω–∏—Ç—å –≤–æ–∑–º–æ–∂–Ω—ã–µ –≤—ã–±—Ä–æ—Å—ã –∏ –ø–æ–¥–æ–±—Ä–∞—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–µ –∞–ª–≥–æ—Ä–∏—Ç–º—ã –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∫–ª—é—á–µ–≤—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

## –§–æ—Ä–º–∞—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è

–†–∞–±–æ—Ç–∞ –±—É–¥–µ—Ç –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ –≤ –≤–∏–¥–µ –Ω–∞–±–æ—Ä–∞ —Ñ–∞–π–ª–æ–≤ –≤ GitHub-—Ä–µ–ø–æ–∑–∏—Ç–æ—Ä–∏–∏:

- –û—Ç–¥–µ–ª—å–Ω—ã–π —Ñ–∞–π–ª —Å –∫–æ–¥–æ–º —Ä–∞–∑–≤–µ–¥–æ—á–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö (EDA).
- –û—Ç–¥–µ–ª—å–Ω—ã–µ —Ñ–∞–π–ª—ã —Å –∫–æ–¥–æ–º –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –∑–∞–¥–∞—á.
- –ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏–π –æ—Ç—á—ë—Ç –≤ —Ñ–æ—Ä–º–∞—Ç–µ PDF –∏–ª–∏ DOC —Å –∏–∑–ª–æ–∂–µ–Ω–∏–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏ –≤—ã–≤–æ–¥–æ–≤.

---

–î–∞–ª–µ–µ –≤ –Ω–æ—É—Ç–±—É–∫–µ –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω –ø–æ–¥—Ä–æ–±–Ω—ã–π EDA ‚Äî –ø–µ—Ä–≤—ã–π –∏ –≤–∞–∂–Ω–µ–π—à–∏–π —ç—Ç–∞–ø –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –ø–æ—Å–ª–µ–¥—É—é—â–µ–≥–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π.


# üìå 1. –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
sns.set(style="whitegrid")
plt.rcParams['figure.dpi'] = 100
```

# üìå 2. –ó–∞–≥—Ä—É–∑–∫–∞ Excel-—Ñ–∞–π–ª–∞

```python
file_path = 'C:\\Users\\Dmitriy\\git\\home_work_MIFI\\coursework\\data\\–î–∞–Ω–Ω—ã–µ_–¥–ª—è_–∫—É—Ä—Å–æ–≤–æ–∏_–ö–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–µ_–ú–û.xlsx'
excel_file = pd.ExcelFile(file_path)

# –ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫–∏–µ –ª–∏—Å—Ç—ã –µ—Å—Ç—å
print("–°–ø–∏—Å–æ–∫ –ª–∏—Å—Ç–æ–≤:", excel_file.sheet_names)
```

# üìå 3. –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏–∑ –ø–µ—Ä–≤–æ–≥–æ –ª–∏—Å—Ç–∞

```python
df = excel_file.parse('Sheet1')
```

# üìå 4. –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –∫–ª—é—á–µ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫

```python
df.rename(columns={
    'IC50, mM': 'IC50',
    'CC50, mM': 'CC50'
}, inplace=True)
```

# üìå 5. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ü–µ–ª–µ–≤—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –∫ float

```python
for col in ['IC50', 'CC50', 'SI']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

# üìå 6. –û–±—â–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö

```python
print("–†–∞–∑–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö:", df.shape)
print("–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:", df.isnull().sum().sum())
print("–î—É–±–ª–∏–∫–∞—Ç—ã:", df.duplicated().sum())
print("\n–¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:\n", df.dtypes.value_counts())
```

# üìå 7. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ü–µ–ª–µ–≤—ã–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º

```python
print("\n–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ IC50, CC50, SI:")
print(df[['IC50', 'CC50', 'SI']].describe())
```

# üìå 8. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è IC50, CC50, SI

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for idx, col in enumerate(['IC50', 'CC50', 'SI']):
    sns.histplot(df[col], kde=True, ax=axes[idx])
    axes[idx].set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ {col}')
plt.tight_layout()
plt.show()
```

# üìå 9. –Ø—â–∏–∫–∏ —Å –≤—ã–±—Ä–æ—Å–∞–º–∏

```python
plt.figure(figsize=(15, 4))
for i, col in enumerate(['IC50', 'CC50', 'SI']):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot: {col}')
plt.tight_layout()
plt.show()
```

# üìå 10. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ (IC50, CC50, SI + –ø–µ—Ä–≤—ã–µ 10 –ø—Ä–∏–∑–Ω–∞–∫–æ–≤)

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
subset_cols = ['IC50', 'CC50', 'SI'] + numerical_cols[:10]
plt.figure(figsize=(12, 8))
sns.heatmap(df[subset_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ (–∫–ª—é—á–µ–≤—ã–µ + –ø–µ—Ä–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏)")
plt.show()
```

# üìå 11. Pairplot –¥–ª—è –∫–ª—é—á–µ–≤—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö (–±–µ–∑ –ø—Ä–æ–ø—É—Å–∫–æ–≤)

```python
clean_df = df[['IC50', 'CC50', 'SI']].dropna()
sns.pairplot(clean_df, kind='reg', diag_kind='hist')
plt.suptitle("–í–∑–∞–∏–º–æ—Å–≤—è–∑—å IC50, CC50 –∏ SI", y=1.02)
plt.show()
```

# üìå –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ—á–∏—â–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –≤ CSV

```python
clean_df_full = df.dropna().reset_index(drop=True)  # —É–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ —Å NaN, –ø–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º –∏–Ω–¥–µ–∫—Å—ã
clean_df_full.to_csv('/content/chem_cleaned_data.csv', index=False)

print("–û—á–∏—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —É—Å–ø–µ—à–Ω–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ —Ñ–∞–π–ª chem_cleaned_data.csv")
```
